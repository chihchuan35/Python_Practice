```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('mosquito_Indicator.csv')
```

```python
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])

df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['dayofweek'] = df['date'].dt.dayofweek

df.head()
```

```python
df.describe()
```

```python
df.hist(figsize=(12, 8))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14,5))
plt.plot(df['date'], df['mosquito_Indicator'])
plt.title("Mosquito Indicator Over Time")
plt.xlabel("Date")
plt.ylabel("Indicator")
plt.show()
```

```python
monthly = df.groupby('month')['mosquito_Indicator'].mean()

plt.figure(figsize=(8,4))
plt.plot(monthly.index, monthly.values)
plt.title("Average Mosquito Indicator by Month")
plt.xlabel("Month")
plt.ylabel("Average Indicator")
plt.show()
```

```python
plt.scatter(df['mean_T(℃)'], df['mosquito_Indicator'], alpha=0.5)
plt.title("Temperature vs Mosquito Indicator")
plt.xlabel("Mean Temperature (°C)")
plt.ylabel("Indicator")
plt.show()
```
